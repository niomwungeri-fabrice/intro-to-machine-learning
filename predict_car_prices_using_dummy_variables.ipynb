```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('carprices.csv')
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('Sell Price($)')
plt.ylabel('Mileage')
plt.scatter(df['Sell Price($)'],df.Mileage, marker='+', color='red')
```

#### Replace Car model with dummy variables (machine learning are good with numbers)

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
new_df = pd.concat([df, dummies], axis='columns')
new_df
```

```python
final_df = new_df.drop(['Car Model', 'Mercedez Benz C class'], axis='columns')
final_df
```

```python
X = final_df.drop(['Sell Price($)'], axis='columns')
X
```

```python
y = final_df['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
# 45000 = Mileage, 4 = age, 0,0 = Mercedez Benz
model.predict([[45000, 4, 0, 0]])
```

```python
# 86000 = Mileage, 7 = age, 1,0 = BMW X5
model.predict([[86000, 7, 1, 0]])
```

```python
### model score
print("{}%".format(round(model.score(X, y) * 100, 2)))
```
